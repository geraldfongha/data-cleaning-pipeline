```python
# Test the data cleaning pipeline
# This function will test the results of the data cleaning pipeline
def test_pipeline_results(cleaned_df):
    # Test: No missing values
    assert cleaned_df.isnull().sum().sum() == 0, "There are still missing values!"


    # Test: No duplicates
    assert cleaned_df.duplicated().sum() == 0, "Duplicates still exist!"

    # Test: No extreme outliers (all values should be within 3 std replaced)
    numeric_cols = cleaned_df.select_dtypes(include=['float64', 'int64']).columns
    for col in numeric_cols:
        m = cleaned_df[col].mean()
        std = cleaned_df[col].std()
        lower_bound = m - 3*std
        upper_bound = m + 3*std
        outlier_count = cleaned_df[(cleaned_df[col] < lower_bound) | (cleaned_df[col] > upper_bound)].shape[0]
        assert outlier_count == 0, f"Outliers still present in column: {col}"

    print("✅ All tests passed!")

# Run test
test_pipeline_results(cleaned_df)
```
